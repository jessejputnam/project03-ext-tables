```python
import pandas as pd
```

```python
# Read in csv
event_cameo = pd.read_csv("cameo.csv", dtype={"code": "str"})

# Get categories from df
event_cameo_cat = event_cameo[event_cameo["code"].str.len() < 3].rename(columns = {"code": "cat_code", "description": "category"})

# Drop category rows from df
event_cameo.drop(
    event_cameo[event_cameo["code"].str.len() < 3].index, 
    inplace=True
)

# Make column of categories per row
event_cameo["cat_code"] = event_cameo["code"].str[:2]

event_cameo_cat

# Join df and return only necessary columns
cols = ["code", "description", "category"]
df = event_cameo.merge(event_cameo_cat, on="cat_code")[cols]

# Set index
df.set_index("code", inplace=True)

# Write to CSV
df.to_csv("../_final/event_cameo.csv")
```
